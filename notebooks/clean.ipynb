```python
import os
import requests
from dotenv import load_dotenv
import time
import numpy as np
load_dotenv()

API_KEY_TMDB = os.getenv("API_KEY_TMDB")

countrys = ["US", "GB", "CA", "FR", "JP", "DE", "ES", "IT", "BR"]

BASE_URL = "https://api.themoviedb.org/3"
HEADERS = {
    "accept": "application/json",
    "Authorization": f"Bearer {API_KEY_TMDB}"
}
```

```python
def _find_id(title: str, year: int, imdb_id=False):

    url = f"{BASE_URL}/search/movie"
    params = {
        "query": title,
        "year": year,
        "language": "en-US",
        "page": 1
    }
    
    try:
        response = requests.get(url, params=params, headers=HEADERS) 
        response.raise_for_status()
        
        data = response.json()
        results = data.get("results", [])
        
        if results:
            tmdb_id = results[0].get("id")
            
            if imdb_id:
                url = f"{BASE_URL}/movie/{tmdb_id}/external_ids"
                response = requests.get(url, headers=HEADERS)
                response.raise_for_status()
                return response.json().get("imdb_id")
            
            return tmdb_id
        else:
            return None
            
    except requests.RequestException as e:
        print(f"  -> Erro ao buscar ID no TMDB para '{title}': {e}")
        return None
```

```python
def _get_release_certificate(movie_id: int):
    url = f"{BASE_URL}/movie/{movie_id}/release_dates"
    try:
        response = requests.get(url, headers=HEADERS)
        response.raise_for_status()
    except requests.RequestException as e:
        print(f"Erro na requisição para o filme ID {movie_id}: {e}")
        return None, None

    results = response.json().get("results", [])
    if not results:
        return None, None

    certification = None
    releases_by_country = {r["iso_3166_1"]: r["release_dates"] for r in results}

    for country in countrys:
        if country in releases_by_country:
            for release in releases_by_country[country]:
                cert = release.get("certification")
                if cert:
                    certification = cert
                    break
        if certification:
            break

    all_dates = [
        release["release_date"]
        for country_releases in releases_by_country.values()
        for release in country_releases
        if release.get("release_date")
    ]

    most_recent_date = max(all_dates).split("T")[0] if all_dates else None

    return certification, most_recent_date
```

```python
def _get_keywords(id_movie):
    url = f"{BASE_URL}/movie/{id_movie}/keywords"

    try:
        response = requests.get(url, headers=HEADERS)
        response.raise_for_status()
        data = response.json()
    
        list_keywords = [i['name'] for i in data.get('keywords', [])]

    except requests.RequestException as e:
        print(f"Erro ao buscar keywords para o filme ID {id_movie}: {e}")
        return []

    return list_keywords

def preencher_dados_faltantes(df: pd.DataFrame, inicio=0) -> pd.DataFrame:
    df_copy = df.copy()
    df_copy['Keywords'] = None

    for index in range(inicio, len(df_copy)):
        title = df_copy.loc[index, "Series_Title"]
        year = df_copy.loc[index, "Released_Year"]

        print(f"\nProcessando: '{title}' ({year})")
        
        time.sleep(0.5)
        try:
            movie_id = _find_id(title, year)
        except Exception as e:
            print(f"Erro ao buscar ID para '{title}': {e}")
            return df_copy, index

        if not movie_id:
            print(f"não encontrou id {title}")
            continue
        

        keywords = _get_keywords(movie_id)
        df_copy.at[index, 'Keywords'] = keywords

        try:
            cert, date = _get_release_certificate(movie_id)
        except Exception as e:
            print(f"Erro ao buscar informações de lançamento para '{title}': {e}")
            return df_copy, index

        if df_copy.at[index, "Certificate"] != cert:
            print(f'discrepancia encontrada em {title}: {df_copy.at[index, "Certificate"]}/{cert}')

        if cert:
            df_copy.at[index, "Certificate"] = cert
            print(f"certificação atualizada: {cert}")

        if date:
            df_copy.at[index, "date_details"] = date
            print(f"data de lançamento atualizada: {date}")

    return df_copy

df_enriched = preencher_dados_faltantes(df)
```

```python
def _get_metascore(title, year, api_key):
    id_movie = _find_id(title, year, imdb_id=True)
    

    url = f"https://imdb236.p.rapidapi.com/api/imdb/{id_movie}/metascore"

    headers = {
        "x-rapidapi-key": api_key,
        "x-rapidapi-host": "imdb236.p.rapidapi.com"
    }
    
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        data = response.json()

        metascore = data['metascore'] if data['metascore'] else None
        
        if metascore:
            print(f'Metascore encontrado: {metascore}')

        return metascore
        
    except requests.RequestException as e:
        print(f"Erro ao buscar Metascore para ID {id_movie}: {e}")
        return 
    except KeyError:
        print(f"Metascore não encontrado para ID {id_movie}")
        return



def put_metascore(df):
    df_copy = df.copy()
    count = 0
    key_api = os.getenv("API_KEY_RAPID_1")
    for i in df_copy[df_copy['Meta_score'].isna()].index:
        count += 1
        if count == 80:
            key_api = os.getenv("API_KEY_RAPID_2")
        title = df_copy.loc[i, 'Series_Title']
        year = df_copy.loc[i, 'Released_Year']

        print(f'Buscando Metascore para {title} ({year})')
        time.sleep(0.5)
        metascore = _get_metascore(title, year, key_api)

        if metascore:
            df_copy.loc[i, 'Meta_score'] = metascore
        


    return df_copy

df_enriched = put_metascore(df_enriched)
df_enriched['Meta_score'].isna().sum()
```

```python
def _get_details_movie(title: str, year: int):

    search_url = f"{BASE_URL}/search/movie"
    params = {
        "query": title,
        "year": year,
        "language": "en-US",
        "page": 1
    }
    
    try:
        response = requests.get(search_url, params=params, headers=HEADERS)
        response.raise_for_status()
        data = response.json()
        results = data.get("results", [])

        if not results:
            print(f"Filme '{title}' ({year}) não encontrado.")
            return None

        tmdb_id = results[0].get("id")



        details_url = f"{BASE_URL}/movie/{tmdb_id}"
        details_response = requests.get(details_url, headers=HEADERS)
        details_response.raise_for_status()
        details = details_response.json()


        return {
            "runtime": details.get("runtime"),
            "original_language": details.get("original_language"),
            "production_companies": [c["name"] for c in details.get("production_companies", [])],
            "production_countries": [c["name"] for c in details.get("production_countries", [])],
            "budget": details.get("budget"),
            "revenue": details.get("revenue"),
            "popularity": details.get("popularity"),


        }

    except requests.RequestException as e:
        print(f"Erro ao buscar '{title}' ({year}): {e}")
        return None



def enrich_financial_data(df: pd.DataFrame) -> pd.DataFrame:
    df_copy = df.copy()

    new_columns = [
        'Budget', 'Revenue', 'Popularity', 
        'Original_Language', 'Production_Companies', 'Production_Countries',

    ]
    for col in new_columns:
        if col not in df_copy.columns:

            df_copy[col] = pd.NA

    total_movies = len(df_copy)

    for index in df_copy.index:
        title = df_copy.loc[index, "Series_Title"]
        year = df_copy.loc[index, "Released_Year"]

        print(f"\n({index + 1}/{total_movies}) Processando: '{title}' ({year})")
        time.sleep(0.5)  

        movie_data = _get_details_movie(title, year)

        if movie_data:
            print(f"  -> Dados encontrados para '{title}'. Atualizando colunas.")
            

            df_copy.at[index, 'Budget'] = movie_data.get('budget')
            df_copy.at[index, 'Revenue'] = movie_data.get('revenue')
            df_copy.at[index, 'Popularity'] = movie_data.get('popularity')
            df_copy.at[index, 'Original_Language'] = movie_data.get('original_language')
            df_copy.at[index, 'Production_Companies'] = movie_data.get('production_companies')
            df_copy.at[index, 'Production_Countries'] = movie_data.get('production_countries')
            
            if pd.isna(df_copy.loc[index, 'Gross']) and movie_data.get('revenue') is not None:
                df_copy.at[index, 'Gross'] = movie_data.get('revenue')
                print(f"  -> Coluna 'Gross' (nula) preenchida com revenue: {movie_data.get('revenue')}")

    print("\nProcesso de enriquecimento de dados concluído.")
    return df_copy


df_enriched = enrich_financial_data(df_enriched)

display(df_enriched)
```
